```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.reuters.com/finance/stocks/company-officers/GOOG.O'
html = requests.get(url)
```

```python
bsObj = BeautifulSoup(html.text,'lxml')
```

```python
bsObj
```

```python
body = bsObj.find('tbody',attrs={'class':'dataSmall'}).getText()
```

```python
body.strip()
body.rstrip()
body = body.replace("\t", ",").replace("\r", ",").replace("\n", ",").replace("\xa0", ",")
```
