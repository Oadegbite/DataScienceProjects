```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,1,1])

ax.plot(x,y,c='red',label='Sine')
ax.plot(x,z,c='green',lw=1,label='Cos')
ax.plot(y,z,c='black',alpha=0.2,label='Circle')

ax.set_xlim(0,10)
ax.set_ylim(-1.5,1.5)

ax.set_xlabel("X linspace")
ax.set_ylabel("Vertical Height")

ax.legend()
ax.set_title("My First Graph")
```

```python
def fun(x):
    return pow(x,2) + -3*x + 8

xArr = list()
yArr = list()

for i in range(101):
    xArr.append(i)
    yArr.append(fun(i))
    
print(xArr)
print(yArr)
```

```python
fig, ay = plt.subplots(nrows=2,ncols=2,figsize=(10,10))

ay[0,0].set_xlabel('Time')
ay[0,0].plot(x,y,c='red',label='red')
ay[1,0].plot(x,z,label='blue')
ay[0,1].set_xlabel('Flat')
ay[0,1].plot(y,z,c='green',label='green')
ay[1,1].plot(x,z,c='purple',label='purple')
fig.legend()
```
